```python
import requests
from bs4 import BeautifulSoup
```

```python
data=requests.get("https://www.airlinequality.com/airline-reviews/british-airways/page/28/")
table=BeautifulSoup(data.text,"html")
table
```

```python
table_name=table.find_all(class_="text_sub_header")
```

```python
table_name
```

```python
two_d_list_name=[]
for data in table_name:
    indiviual_name=data.find_all(itemprop="name")
    customer_name=[data.text.strip() for data in indiviual_name]
    two_d_list_name.append(customer_name)
print(two_d_list_name)

```

```python
two_d_list_customerfeedback=[]
table_rating=table.find_all(class_="text_content",itemprop="reviewBody")
# table_rating
for data in table_rating:
    customer_feedback=[data.text.strip()]
    two_d_list_customerfeedback.append(customer_feedback)
print(two_d_list_customerfeedback)

for data in two_d_list_customerfeedback:
    print(data)
```

```python

# Remove prefixes from each string
cleaned_reviews = [review[0].replace('â Trip Verified | ', '').replace('Not Verified | ', '') for review in two_d_list_customerfeedback]

# Print the cleaned reviews
for cleaned_review in cleaned_reviews:
    print(cleaned_review)
```

```python
rating_table=table.find_all(class_="text_header")
rating_table
```

```python
two_d_list_crisp_customer_feedback=[]
for data in rating_table:
    customer_feedback=[data.text.strip()]
    two_d_list_crisp_customer_feedback.append(customer_feedback)
for data in two_d_list_crisp_customer_feedback:
    print(data)
```

```python
two_d_list_rating=[]
rating=table.find_all(class_="rating-10")
for data in rating[1:]:
  indiviual_rating=data.find_all(itemprop="ratingValue")
  indiviual_rating1=[data.text.strip() for data in indiviual_rating]
  two_d_list_rating.append(indiviual_rating1)
print(two_d_list_rating)
```

```python
customer_rating_list = [[int(num) for num in row] for row in two_d_list_rating]
```

```python
cols1=["Name"]
cols2=["Rating(10points)"]
cols3=["Crisp_Feedback"]
cols4=["Full_Descripation"]

import pandas as pd
```

```python
data1=two_d_list_name
data2=customer_rating_list
data3=two_d_list_crisp_customer_feedback
data4=cleaned_reviews
df1=pd.DataFrame(data1,columns=cols1)
df2=pd.DataFrame(data2,columns=cols2)
df3=pd.DataFrame(data3,columns=cols3)
df4=pd.DataFrame(data4,columns=cols4)


result_df = pd.concat([df1, df2,df3,df4], axis=1)
result_df.head(9)
```
